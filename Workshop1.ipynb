```python
# Display Fibonacci Series

n = 10
a, b = 0,1

print("Fibonacci Series:")
for i in range(n):
  print(a, end = " ")
  a, b = b, a+b
```

```python
# Factorial of a number

num = int(input("Enter a number:"))

factorial = 1

if num < 0:
  print("Fsctorial does not exist for negative numbers.")
elif num == 0:
  print ("Factorial of 0 is 1")
else:
  for i in range (1, num+1):
    factorial *= i
  print (f"factorial of {num} is {factorial}")
```

```python
# Print right angled triangle using for loop

rows = 5

for i in range(1, rows+1):
  print("*" * i)
```

```python
# Simple calculator using function

def add(a, b):
  return a+b

def subtract(a, b):
  return a-b

def multiply(a, b):
  return a*b

def divide(a, b):
  if b == 0:
    return "Cannot divide by zero"
  return a / b

# Taking input

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition: ", add(num1, num2))
print("Subtraction:", subtract(num1, num2))
print("Multiplication:", multiply(num1, num2))
print("Division:", divide(num1, num2))
```

```python
# Remove the First Duplicate Element of a List

# Remove first duplicate element

my_list = [1, 2, 3, 2, 4, 5]

seen = set()

for i in range(len(my_list)):
    if my_list[i] in seen:
        my_list.pop(i)
        break
    seen.add(my_list[i])

print("Updated List:", my_list)
```

```python
# Create New Dictionary by Extracting Mentioned Keys

sample_dict = {
    "name": "Anderson",
    "age": 35,
    "salary": 5000,
    "city": "London"
}

keys = ["name", "salary"]

new_dict = {k: sample_dict[k] for k in keys}

print(new_dict)
```

```python
# Convert Two Lists into a Dictionary

list_one = ['one', 'two', 'three']
list_two = [1, 2, 3]

result = dict(zip(list_one, list_two))

print(result)
```
